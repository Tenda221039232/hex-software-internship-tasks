```python
import matplotlib.pyplot as plt

df.groupby("sex")[['G1', 'G2', 'G3']].mean().mean(axis=1).sort_values(ascending=False).plot(kind='pie', autopct='%1.1f%%')

plt.title("Different Genders' Performance")
plt.ylabel("")  # Optional: removes y-axis label for a cleaner pie chart
plt.show()
```
